```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import oscovida.plots._plotly as plots
```

```python
#  Used to convert plotly output to PNG
from IPython.display import Image
```

```python
region = pd.read_csv(
    "../test_data/20200901_level_1.csv",
    index_col='date',
    parse_dates=['date'],
)
```

```python
data = region[region.id=='DEU']
```

```python
fig = plots.plot_totals(data.confirmed)
Image(plots.plot_totals(data.deaths, fig).to_image(format="png"))
```

```python
fig = plots.plot_totals(data.confirmed, label_prepend="DEU")
Image(plots.plot_totals(data.deaths, fig, label_prepend="DEU").to_image(format="png"))
```
